```python
#reading the data 
#FIRST OF ALLL
#read headers
print(df.columns)
```

```python
#using loc to get based on some condition 
df.loc[df['Type 1']=="Fire"]
```

```python
df.describe()
```

```python
#SORTING N DESCRIBING DATA
df.sort_values('Name')
```

```python
#more specificly
df.sort_values(["Type 1","HP"])
```

```python
#MAKING CHANGES TO DATA 
#If i want a one column with all these stats ..to get which pokemon is more powerful we gonna do
df["total"]=df["HP"]+df["Sp. Def"]+df["Sp. Atk"]+df["Speed"] +df["Defense"]
df
```

```python
#if you wanna delete some columns 
df=df.drop(columns=['total','HP','Generation'])
df
```

```python
#Another way to do total
df["total"]=df.iloc[:,4:10].sum(axis=1)
df
```

```python
#to chnage position of columns 
cols=list(df.columns.values)
df= df[cols[0:3]+[cols[-1]]+cols[3:10]]
df
```

```python
df
```

```python
#FILTERING DATA
df.loc[df["Type 1"]=="Grass"]
```

```python
#GIVING MULTIPLE FILTERS
df.loc[(df["Type 1"]=="Grass") & (df["Type 2"]=="Poison")]  # can also do "OR"
```

```python
#if i want all pokemon with total greater than 300
df.loc[(df["total"]>=300)]
```

```python
#some more conditions
df.loc[(df["Type 1"]=="Grass") & (df["Type 2"]=="Poison") & (df["Defense"]>80)]
```

```python
#FILTERATION BASED ON TEXTUAL PATTRENS
df.loc[df["Name"].str.contains('saur')]
```

```python
#CONDITIONAL CHANGES
#I want flamer instaed of fire in type 1
df.loc[df["Type 1"]=="Fire","Type 1"]="Flamer"
df
```

```python
df.loc[df["total"]>300,"Legendary"]="GOOD"
df
```

```python
#AGGREGATE STATISTICS (GROUP BY)
df.groupby(["Type 1"]).mean()
```

```python
#if i want to count
df.groupby(["Type 1"]).count()
```

```python
#WRITING DATFRAME BACK INTO CSV 
df.to_csv("newmodified.csv")
```
